# 타 데이터와 MERGE를 위해 '년도','산업분류','시총' 순으로 DF 재구성

```python
# 코스피, 코스닥 시총 통합 데이터 다시 불러오기
mcz_by_industry = pd.read_csv('../data/mcz_by_industry.csv', encoding='euc-kr')
mcz_by_industry.head()
```

```python
# 'Unnamed: 0' 컬럼 제거
mcz_by_industry.drop(columns=['Unnamed: 0'], inplace=True)
mcz_by_industry.head()
```

```python
# '년도','산업분류','시총' 순으로 데이터 프레임 재구성
reform_df = pd.DataFrame(columns=['년도','산업분류','시총'])
```

```python
# 기존 'mcz_by_industry' 통합 데이터로 'reform_df'에 추가
for idx, title in enumerate(mcz_by_industry.columns[1:]) :
    df_tmp = pd.DataFrame(columns=['년도','산업분류','시총'])
    df_tmp['년도'] = mcz_by_industry.iloc[:,0]
    df_tmp['산업분류'] = title
    df_tmp['시총'] = mcz_by_industry.iloc[:,idx+1]
    reform_df = reform_df.append(df_tmp)
```
