```python
df['년도'] = industry_sum.iloc[:,0]
df['산업분류'] = industry_sum.columns[1]
df['시총'] = industry_sum.iloc[:,1]
df
```

```python
for idx, title in enumerate(industry_sum.columns[1:]) :
    df_tmp = pd.DataFrame(columns=['년도','산업분류','시총'])
    df_tmp['년도'] = industry_sum.iloc[:,0]
    df_tmp['산업분류'] = title
    df_tmp['시총'] = industry_sum.iloc[:,idx+1]
    df = df.append(df_tmp)
```
